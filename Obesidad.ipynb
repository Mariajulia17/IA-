Se estará comparando la variable de estatura con el nivel de obesidad para el sexo femenino, esto ya que al investigar factores para identificar la obesidad mencionaban mucho la estatura como un rasgo importante, por lo que quiero ver la relación que existen entre estas variables y ver el nivel de obesidad que puede llegar a tener en esta población.

```python
#Función mágica, sirve para que las imágenes se desplieguen sin problema en las *Jupyter Notebooks*
%matplotlib inline
import matplotlib.pyplot as plt
plt.scatter(x = data.Estatura[data.Sexo=="Female"], y = data.NivelDeObesidad[data.Sexo=="Female"], alpha = 0.05)
plt.show()
```
